### 1) Trade Forecasts - Guangdong Province
Using a long Short-Term Memory (LSTM) Model, we calculated prediction of exports of Guandong Province using monthly trade data for 6 periods. The first drilldown defined on **drilldowns** must be a time level with **parents=true**.

API:
```
https://dev.oec.world/api/stats/prophet?Subnat Geography=44&cube=trade_s_chn_m_hs&drilldowns=Month&measures=Trade Value&parents=true&sparse=false&periods=6&Trade Flow=2
```
